```python
import yfinance as yf
import csv
from datetime import datetime
import time

#store historical commodity data
csv_file = "commodities_prices.csv"

#commodities to track
commodities = {
    "Gold": "GC=F",
    "Silver": "SI=F",
    "Crude Oil": "CL=F",
    "Natural Gas": "NG=F",
    "Copper": "HG=F"
}

#fetch every 12 hours
fetch_interval = 12 * 3600

while True:
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    row = [timestamp]
    
    print(f"Fetching commodity prices at {timestamp}...")
    for name, symbol in commodities.items():
        try:
            price = yf.Ticker(symbol).info["regularMarketPrice"]
            print(f"{name}: {price}")
            row.append(price)
        except Exception as e:
            print(f"Error fetching {name}: {e}")
            row.append(None)

    #save
    with open(csv_file, mode="a", newline="") as f:
        writer = csv.writer(f)
        if f.tell() == 0:
            #write header for empty file
            writer.writerow(["Timestamp"] + list(commodities.keys()))
        writer.writerow(row)

    print(f"Data saved. Waiting {fetch_interval / 3600} hours until next fetch...\n")
    time.sleep(fetch_interval)

```
